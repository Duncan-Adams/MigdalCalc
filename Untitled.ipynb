```python
import os

import numpy as np
from scipy.optimize import fsolve

from migdalcalc import migdalcalc as mc
from migdalcalc.migdal import quench
from migdalcalc.migdal import ionization as ion
from migdalcalc.migdal import kinematics as kin

%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
plt.rcParams['figure.dpi'] = 100
plt.rcParams['text.usetex'] = False
```

```python
Si_migdal = mc.migdalcalc(target="Si-28", nuclear_data_name="jeff33")
```

```python
de_range = np.geomspace(1, 1e3)
```

```python
plt.plot(de_range, Si_migdal.mig.dP_dDeltaE(de_range,method="elf"),label="elf")
plt.plot(de_range, Si_migdal.mig.dP_dDeltaE(de_range,method="ibe"),label="ibe")
plt.yscale('log')
plt.xscale('log')
plt.ylabel("Differential Probability")
plt.xlabel("Electron Energy (eV)")
plt.legend()
plt.show()
```
