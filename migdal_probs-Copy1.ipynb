```python
from migdalcalc import migdalcalc
from migdalcalc.migdal import kinematics as kin
import matplotlib.pyplot as plt
import numpy as np
import scipy.integrate as integrate
```

```python
Si_migdal = migdalcalc.migdalcalc(target='Si-28')
Xe_migdal = migdalcalc.migdalcalc(target='Xe-131')
Ar_migdal = migdalcalc.migdalcalc(target="Ar-40")
```

```python
dE_range = np.geomspace(1, 100)
plt.plot(Si_migdal.mig.dP_dDeltaE(dE_range, 'elf'), label='elf')
plt.plot(Si_migdal.mig.dP_dDeltaE(dE_range, 'ibe'), label='ibe')
plt.legend()
plt.yscale('log')
plt.show()
```

```python
dE_range = np.geomspace(1, 1e5)

plt.plot(dE_range, Xe_migdal.mig.dP_dDeltaE(dE_range, 'ibe'))
plt.yscale('log')
plt.xscale('log')
plt.title("Xe Migdal Ionization Probability (q = 1 eV)")
plt.show()
plt.cla()

plt.plot(dE_range, Ar_migdal.mig.dP_dDeltaE(dE_range, 'ibe'))
plt.yscale('log')
plt.xscale('log')
plt.title("Ar Migdal Ionization Probability (q = 1 eV)")
plt.show()
```
